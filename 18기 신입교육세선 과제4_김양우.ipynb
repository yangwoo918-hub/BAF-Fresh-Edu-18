```python
def missing_table(df):
    # 각 컬럼별 결측치 개수
    miss = df.isna().sum()
    # 결측치가 1개 이상인 컬럼만 남기면
    miss = miss[miss > 0].sort_values(ascending=False)
    return pd.DataFrame({
        "missing_count": miss,
        "missing_rate(%)": (miss / len(df) * 100).round(2)
    })

print("[결측치 요약 - train]")
display(missing_table(train_raw))

print("[결측치 요약 - test]")
display(missing_table(test_raw))
```

```python
def _gamma_from_dewpoint(Td):
    #    Td(이슬점)에서 γ를 계산하는 부분 γ = (17.62 * Td) / (243.12 + Td)
    Td = np.asarray(Td, dtype=float)
    return (17.62 * Td) / (243.12 + Td)

def dewpoint_from_temp_rh(T, RH):
    #기온 T와 습도 RH로 이슬점 Td 계산
    T = np.asarray(T, dtype=float)
    RH = np.asarray(RH, dtype=float)

    # RH가 0이면 계산 불가능 → 최솟값으로
    RH = np.clip(RH, 0.1, 100.0)

    gamma = np.log(RH / 100.0) + (17.62 * T) / (243.12 + T)
    Td = (243.12 * gamma) / (17.62 - gamma)
    return Td

def temp_from_dewpoint_rh(Td, RH):
    #역변환으로 Td와 RH로 기온 T 계산 - 결측 기온을 채우기 위해 사용
    
    Td = np.asarray(Td, dtype=float)
    RH = np.asarray(RH, dtype=float)
    RH = np.clip(RH, 0.1, 100.0)

    gamma = _gamma_from_dewpoint(Td)

    # γ = ln(RH/100) + 17.62*T/(243.12+T)
    # 17.62*T/(243.12+T) = γ - ln(RH/100) = A 라고 두면
    A = gamma - np.log(RH / 100.0)

    # A = 17.62*T/(243.12 + T) 를 T에 대해 풀면 아래 식
    T = (243.12 * A) / (17.62 - A)
    return T

def rh_from_temp_dewpoint(T, Td):
    
    T = np.asarray(T, dtype=float)
    Td = np.asarray(Td, dtype=float)
    gamma = _gamma_from_dewpoint(Td)
    ln_rh = gamma - (17.62 * T) / (243.12 + T)
    RH = 100.0 * np.exp(ln_rh)
    return np.clip(RH, 0.1, 100.0)
```

```python
#이슬점 공식을 역이용해 온도 결측치 처리
#결측치가 없는 온도 행만 추출
mask_ok = train_raw["Temperature(째C)"].notna() & (train_raw["Humidity(%)"] > 0)
#T_est는 역변환으로 저장한 기온
T_est = temp_from_dewpoint_rh(
    train_raw.loc[mask_ok, "Dew point temperature(째C)"],
    train_raw.loc[mask_ok, "Humidity(%)"]
)
#실제 기온과 추정값의 오차의 평균->0.0676(차이가 거의없음)
mae = np.mean(np.abs(T_est - train_raw.loc[mask_ok, "Temperature(째C)"]))
print(f"[검증] 이슬점+습도 -> 기온 역변환 MAE(°C): {mae:.4f}")
```

```python


def month_to_season(m):
    if m in [12, 1, 2]: return "Winter"
    if m in [3, 4, 5]: return "Spring"
    if m in [6, 7, 8]: return "Summer"
    return "Autumn"

def preprocess(df):
    df = df.copy()  

   
    #Date 처리 + 파생변수 생성
    
    df["Date"] = pd.to_datetime(df["Date"], format="%d/%m/%Y", errors="coerce")
    df["Year"] = df["Date"].dt.year
    df["Month"] = df["Date"].dt.month
    df["Day"] = df["Date"].dt.day
    df["DayOfWeek"] = df["Date"].dt.dayofweek     # 월=0,화=1... 일=6
    df["IsWeekend"] = (df["DayOfWeek"] >= 5).astype(int)  # 토/일이면 1

    
    #Seasons 결측 -> Month 기준으로 복원
   
    df["Seasons"] = df["Seasons"].fillna(df["Month"].map(month_to_season))

    
    #Humidity(%)에서 0은 이상치 가능성이 높으므로 NaN 처리
    
    df["Humidity_clean"] = df["Humidity(%)"].replace(0, np.nan).astype(float)

    
    #Td + RH로 Temperature 결측치 채우기
    
    # Td와 RH로 계산한 추정 기온 생성
    df["Temp_from_dew_RH"] = temp_from_dewpoint_rh(
        df["Dew point temperature(째C)"],
        df["Humidity_clean"]
    )

    # Temperature를 채울 새 컬럼
    df["Temperature_filled"] = df["Temperature(째C)"].copy()

    # Temperature가 결측이고, 추정 기온이 존재하면 그 값으로 채움
    cond = df["Temperature_filled"].isna() & df["Temp_from_dew_RH"].notna()
    df.loc[cond, "Temperature_filled"] = df.loc[cond, "Temp_from_dew_RH"]

    # 그래도 남는 Temperature 결측치는 그룹 중앙값으로 채움
    # 1순위: (Seasons, Month, Hour) 기준 중앙값
    df["Temperature_filled"] = df["Temperature_filled"].fillna(
        df.groupby(["Seasons", "Month", "Hour"])["Temperature_filled"].transform("median")
    )
    # 2순위: (Seasons, Hour)
    df["Temperature_filled"] = df["Temperature_filled"].fillna(
        df.groupby(["Seasons", "Hour"])["Temperature_filled"].transform("median")
    )
    # 3순위: 전체 중앙값
    df["Temperature_filled"] = df["Temperature_filled"].fillna(df["Temperature_filled"].median())

    # ---------------------------
    #습도 결측도 T_filled와 Td로 복원
    # ---------------------------
    df["RH_from_T_Td"] = rh_from_temp_dewpoint(
        df["Temperature_filled"],
        df["Dew point temperature(째C)"]
    )
    df["Humidity_filled"] = df["Humidity_clean"].fillna(df["RH_from_T_Td"])

    
    #Wind speed 결측은 그룹 중앙값으로 채우기
    df["Wind speed (m/s)"] = df["Wind speed (m/s)"].fillna(
        df.groupby(["Seasons", "Hour"])["Wind speed (m/s)"].transform("median")
    )
    df["Wind speed (m/s)"] = df["Wind speed (m/s)"].fillna(
        df.groupby(["Seasons"])["Wind speed (m/s)"].transform("median")
    )
    df["Wind speed (m/s)"] = df["Wind speed (m/s)"].fillna(df["Wind speed (m/s)"].median())

    #모델링에 편하도록 이진 변수도 추가
    df["Holiday_bin"] = (df["Holiday"] == "Holiday").astype(int)
    df["FunctioningDay_bin"] = (df["Functioning Day"] == "Yes").astype(int)

    return df
```

```python
RANDOM_STATE = 42


train_model = train.copy()
test_model  = test.copy()


train_model["Temperature(째C)"] = train_model["Temperature_filled"]
test_model["Temperature(째C)"]  = test_model["Temperature_filled"]

train_model["Humidity(%)"] = train_model["Humidity_filled"]
test_model["Humidity(%)"]  = test_model["Humidity_filled"]


drop_tmp_cols = [
    "Temperature_filled", "Humidity_filled",
    "Humidity_clean", "Temp_from_dew_RH", "RH_from_T_Td"
]
train_model.drop(columns=drop_tmp_cols, inplace=True, errors="ignore")
test_model.drop(columns=drop_tmp_cols,  inplace=True, errors="ignore")


# Date는 제거

if "Date" in train_model.columns:
    train_model.drop(columns=["Date"], inplace=True)
if "Date" in test_model.columns:
    test_model.drop(columns=["Date"], inplace=True)


train_model = train_model.replace([np.inf, -np.inf], np.nan)
test_model  = test_model.replace([np.inf, -np.inf], np.nan)

print("train_model 결측치 총합:", train_model.isna().sum().sum())
print("test_model  결측치 총합:", test_model.isna().sum().sum())


#X, y 분리

y = train_model["Rented Bike Count"]
X = train_model.drop(columns=["Rented Bike Count"])

display(X.head())
display(y.head())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_STATE
)

print("X_train:", X_train.shape, "X_valid:", X_valid.shape)
print("y_train:", y_train.shape, "y_valid:", y_valid.shape)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

def eval_regression(model, X_train, y_train, X_valid, y_valid, name="model"):
    """학습/검증 성능을 RMSE/MAE/R2로 출력"""

    pred_tr = model.predict(X_train)
    pred_va = model.predict(X_valid)

    rmse_tr = np.sqrt(mean_squared_error(y_train, pred_tr))
    rmse_va = np.sqrt(mean_squared_error(y_valid, pred_va))

    mae_tr = mean_absolute_error(y_train, pred_tr)
    mae_va = mean_absolute_error(y_valid, pred_va)

    r2_tr = r2_score(y_train, pred_tr)
    r2_va = r2_score(y_valid, pred_va)

    print(f"\n[{name}]")
    print("학습 성능 : RMSE={:.3f} | MAE={:.3f} | R2={:.3f}".format(rmse_tr, mae_tr, r2_tr))
    print("검증 성능 : RMSE={:.3f} | MAE={:.3f} | R2={:.3f}".format(rmse_va, mae_va, r2_va))

    return rmse_va
```

```python
from sklearn.compose import ColumnTransformer, TransformedTargetRegressor
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer

#수치형/범주형 분리
num_features = X.select_dtypes(include=[np.number]).columns.tolist()
cat_features = X.select_dtypes(exclude=[np.number]).columns.tolist()

print("수치형 개수:", len(num_features))
print("범주형 개수:", len(cat_features), cat_features)

#수치형:결측->중앙값 
num_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

#범주형:결측->최빈값 채움
cat_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocess_model = ColumnTransformer(
    transformers=[
        ("num", num_transformer, num_features),
        ("cat", cat_transformer, cat_features)
    ]
)
```

```python
from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor, HistGradientBoostingRegressor

models = {
    #선형회귀
    "Ridge(규제 선형회귀)": Ridge(alpha=5.0),

    # Randomforest
    "RandomForest": RandomForestRegressor(
        n_estimators=300,
        min_samples_leaf=2,
        random_state=RANDOM_STATE,
        n_jobs=-1
    ),

    # Gradientboosting
    "HistGradientBoosting": HistGradientBoostingRegressor(
        learning_rate=0.08,
        max_depth=8,
        random_state=RANDOM_STATE
    )
}

valid_scores = {}
fitted_models = {}

for name, base_model in models.items():
    pipe = Pipeline(steps=[
        ("prep", preprocess_model),
        ("model", base_model)
    ])

    #학습은 log로 하여 오차의 분산을 줄여서 학습하고 예측량은 exp로 다시 되돌려놓음
    model = TransformedTargetRegressor(
        regressor=pipe,
        func=np.log1p,
        inverse_func=np.expm1
    )

    model.fit(X_train, y_train)
    rmse_va = eval_regression(model, X_train, y_train, X_valid, y_valid, name=name)

    valid_scores[name] = rmse_va
    fitted_models[name] = model

best_name = min(valid_scores, key=valid_scores.get)
best_model = fitted_models[best_name]

print("\nBest Model:", best_name)
print("Best Valid RMSE:", valid_scores[best_name])
#RMSE: 평균적으로 몇 대 정도 틀리냐의 지표(MAE보다 오차에 민감)
#R2:0~1,0.8이면 대여량 변동의 80%를 잘 설명
#MAE:평균적으로 예측이 실제값에서 몇 대 정도 틀리나?
```

```python
best_model.fit(X, y)

# test 예측
X_test_final = test_model.replace([np.inf, -np.inf], np.nan)  
test_pred = best_model.predict(X_test_final)

# 대여량은 0 이상이므로 음수=0
test_pred = np.clip(test_pred, 0, None)


submission = pd.DataFrame({
    "Date": test_raw["Date"],
    "Hour": test_raw["Hour"],
    "Rented Bike Count": test_pred
})

display(submission.head())


output_path = "test___predictions.csv"
submission.to_csv(output_path, index=False, encoding="utf-8-sig")
print("저장:", output_path)
```
