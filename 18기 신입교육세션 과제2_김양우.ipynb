## **파생변수 만들기**

## **SibSp, Parch 합치기 -> family 변수 생성**

```python
data1["family"]=data1["SibSp"]+data1["Parch"]
```

```python
# 반드시 확인하기
data1[['family',"SibSp","Parch"]]
```

```python
from scipy.stats import chi2_contingency, mannwhitneyu, ttest_ind

df = data1.copy()

# (전처리용 파생변수들) — 여기서는 분석용으로만 만듦
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1
df["IsAlone"] = (df["FamilySize"] == 1).astype(int)
df["FamilyGroup"] = pd.cut(df["FamilySize"], bins=[0, 1, 4, 20], labels=["Alone(1)", "Small(2-4)", "Large(5+)"])

df["CabinKnown"] = df["Cabin"].notna().astype(int)
df["Deck"] = df["Cabin"].astype(str).str[0]
df.loc[df["Cabin"].isna(), "Deck"] = "U"
df["Deck"] = df["Deck"].replace({"T": "U"})

def cramers_v(ct):
    chi2, p, dof, exp = chi2_contingency(ct)
    n = ct.values.sum()
    r, k = ct.shape
    return np.sqrt(chi2 / (n * (min(r-1, k-1))))

# 범주형 변수들: 카이제곱 + Cramer's V
cat_cols = ["Sex", "Pclass", "Embarked", "CabinKnown", "Deck", "IsAlone", "FamilyGroup"]
rows = []

for col in cat_cols:
    tmp = df.dropna(subset=[col])
    ct = pd.crosstab(tmp[col], tmp["Survived"])
    chi2, p, dof, exp = chi2_contingency(ct)
    v = cramers_v(ct)
    surv_rate = tmp.groupby(col)["Survived"].mean()
    rows.append([col, chi2, p, v, surv_rate.to_dict()])

cat_test = pd.DataFrame(rows, columns=["변수", "chi2", "p-value", "CramersV", "생존률(범주별)"])
display(cat_test.sort_values("p-value"))

# 연속형: Age(결측 제외), Fare
age_df = df.dropna(subset=["Age"])
age_surv = age_df[age_df["Survived"]==1]["Age"]
age_dead = age_df[age_df["Survived"]==0]["Age"]

t_p_age = ttest_ind(age_surv, age_dead, equal_var=False).pvalue
u_p_age = mannwhitneyu(age_surv, age_dead, alternative="two-sided").pvalue

fare_surv = df[df["Survived"]==1]["Fare"]
fare_dead = df[df["Survived"]==0]["Fare"]
u_p_fare = mannwhitneyu(fare_surv, fare_dead, alternative="two-sided").pvalue

print("[Age] 생존자 평균:", round(age_surv.mean(),2), "사망자 평균:", round(age_dead.mean(),2))
print("[Age] t-test p-value:", t_p_age, "| Mann-Whitney p-value:", u_p_age)

print("[Fare] 생존자 중앙값:", round(fare_surv.median(),2), "사망자 중앙값:", round(fare_dead.median(),2))
print("[Fare] Mann-Whitney p-value:", u_p_fare)
```

```python
df_prep = data1.copy()

# Title 추출
df_prep["Title"] = df_prep["Name"].str.extract(r",\s*([^\.]+)\.", expand=False).str.strip()

# Title 정리
df_prep["Title"] = df_prep["Title"].replace({"Mlle": "Miss", "Ms": "Miss", "Mme": "Mrs"})

# 희귀 타이틀은 Rare로 묶기
rare_titles = ["Dr", "Rev", "Major", "Col", "Capt", "Sir", "Lady", "Don", "Jonkheer", "the Countess"]
df_prep["Title"] = df_prep["Title"].replace(rare_titles, "Rare")

print("Age 결측(전):", df_prep["Age"].isna().sum())

(Title, Sex, Pclass) 그룹 중앙값으로 채우기
df_prep["Age"] = df_prep["Age"].fillna(df_prep.groupby(["Title", "Sex", "Pclass"])["Age"].transform("median"))

# 혹시 남는 결측은 (Sex, Pclass) 중앙값 → 그래도 남으면 전체 중앙값
df_prep["Age"] = df_prep["Age"].fillna(df_prep.groupby(["Sex", "Pclass"])["Age"].transform("median"))
df_prep["Age"] = df_prep["Age"].fillna(df_prep["Age"].median())

print("Age 결측(후):", df_prep["Age"].isna().sum())

# Title 분포도 확인(전처리에 유용)
display(df_prep["Title"].value_counts())
```

```python
df_prep["Embarked"] = df_prep["Embarked"].fillna(df_prep["Embarked"].mode()[0])

# Cabin 파생변수
df_prep["CabinKnown"] = df_prep["Cabin"].notna().astype(int)
df_prep["Deck"] = df_prep["Cabin"].astype(str).str[0]
df_prep.loc[df_prep["Cabin"].isna(), "Deck"] = "U"
df_prep["Deck"] = df_prep["Deck"].replace({"T": "U"})



df_prep["Age"] = df_prep["Age"].round().astype(int)   # 반올림 후 int
df_prep["Age"] = df_prep["Age"].clip(lower=1)         # 1 이상(자연수) 보장

# 가족 파생변수
df_prep["family"] = df_prep["SibSp"] + df_prep["Parch"]
df_prep["FamilySize"] = df_prep["family"] + 1
df_prep["IsAlone"] = (df_prep["FamilySize"] == 1).astype(int)

# 티켓 그룹 크기
df_prep["TicketGroupSize"] = df_prep.groupby("Ticket")["Ticket"].transform("count")

# Fare 로그변환
df_prep["Fare_log"] = np.log1p(df_prep["Fare"])

# 모델링에 안 쓸 텍스트/큰 결측 컬럼 제거
drop_cols = ["Name", "Ticket", "Cabin"]
X = df_prep.drop(columns=["Survived"] + drop_cols)
y = df_prep["Survived"]

# 범주형 원-핫 인코딩
cat_cols = ["Sex", "Embarked", "Deck", "Title"]
X_encoded = pd.get_dummies(X, columns=cat_cols, drop_first=True)

print("결측치 총합:", X_encoded.isna().sum().sum())
display(X_encoded.head())


print("Age dtype:", X_encoded["Age"].dtype)
print("Age 소수 존재 여부:", (X_encoded["Age"] % 1 != 0).any())


df_final = pd.concat([y, X_encoded], axis=1)
df_final.to_csv("titanic_preprocessed.csv", index=False)
print("저장 완료: titanic_preprocessed.csv")
```
